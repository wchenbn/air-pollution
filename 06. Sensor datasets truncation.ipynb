```python
import pandas as pd
import numpy as np 
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

### Miladinovci

```python
df = pd.read_csv('./data/formatted-sensor/Miladinovci', index_col = 0)
```

```python
df.head()
```

```python
df = df.drop(['PM25'], axis=1)
```

```python
df_truncated = df.loc['2016-04-14 12:00:00':]
df_truncated.head(24)
```

```python
df_truncated.shape
```

```python
plt.figure(figsize=(17, 8))
sns.heatmap(df_truncated.isnull(), cbar=False).set_title('Miladinovci', fontsize=20)
plt.show()
```

```python
df_truncated.isnull().sum(axis=0)
```

```python
df_truncated.to_csv('./data/truncated-sensor/Miladinovci', index=True)
```
